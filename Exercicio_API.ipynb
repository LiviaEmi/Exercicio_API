```python
# importando biblioteca
import requests

# pegando a url do site Open Meteo
url = 'https://api.open-meteo.com/v1/forecast'

# estabelecendo que as variáveis serão lat e longi, do tipo float
# colocando um texto para informar a latitude e longitude
lat = float(input('Informe a latitude: '))
longi = float(input('Informe a longitude: '))

# definindo os parâmetros, lembrando que eles devem estar do mesmo jeito que no documento da api
parametros = {
    'latitude': lat,
    'longitude': longi,
    'daily': ['temperature_2m_max', 'precipitation_sum'],
    'timezone': 'auto',
    'past_days': 6,
    'forecast_days': 1
    }
# o "past_days" já traz os dados de 6 dias atrás
# o forecast_days traz os dados futuros (e de hoje)
# como o ex pedia 7 dias, defini para trazer 6 dias em past_days
# e o dia atual (1) em forecast_days

# requisição - coletando dados do sistema/internet:
resposta = requests.get(url, params=parametros)

# verificar se está correto:
if resposta.status_code == 200:
# se sim: converter os dados na forma que o python entende (json)
  dados = resposta.json()
else:
  print('Erro ao acessar a API')

# seleção: pega só a parte específica dos dados diários
dados_diarios = dados['daily']

# importando a biblioteca pandas, para depois converter em csv:
import pandas as pd

# transformando os dados em dataframe para poder converter em csv:
df_dados_clima = pd.DataFrame(dados_diarios)
# ^ DataFrame organiza os dados em tabela, esse passo é importante para depois gerar
# uma tabela com colunas e linhas certinha: print(df_dados_clima)
# se colocasse apenas print(dados_clima), iria ficar tudo bagunçado

# exportando em formato csv:
df_dados_clima.to_csv('atividade_clima_Livia.csv', index = False)

# se eu quiser que apareça os dados na tela em formato de tabela bonitinha:
print('      ------------ Histórico do clima ------------')
print(df_dados_clima)

print('\n')
print('Arquivo csv gerado com sucesso')
```
